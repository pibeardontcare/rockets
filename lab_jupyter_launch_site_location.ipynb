```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

from math import sin, cos, sqrt, atan2, radians

# Function to calculate Haversine distance
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Approximate radius of Earth in km

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])  # Convert to radians

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c  # Distance in km

# Define launch site coordinates (Example: CCAFS SLC-40)
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# Define a set of possible coastline coordinates near the launch site
coastlines = [
    {"lat": 28.56367, "lon": -80.57163},  # Closest known coastline
    {"lat": 28.56230, "lon": -80.56700},  # Another nearby coastline
    {"lat": 28.56000, "lon": -80.56000}   # Further away coastline
]

# Find the closest coastline
closest_coastline = min(coastlines, key=lambda c: calculate_distance(launch_site_lat, launch_site_lon, c["lat"], c["lon"]))

# Calculate distance to closest coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, closest_coastline["lat"], closest_coastline["lon"])

# Display results
print(f"Closest Coastline Coordinates: Lat {closest_coastline['lat']}, Lon {closest_coastline['lon']}")
print(f"Distance to Closest Coastline: {distance_coastline:.2f} km")
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )


import folium
from folium.features import DivIcon

# Create a folium map centered at the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Add a marker for the launch site
folium.Marker(
    [launch_site_lat, launch_site_lon], 
    popup="Launch Site",
    icon=folium.Icon(color="blue", icon="rocket")
).add_to(site_map)

# Add a marker for the closest coastline
folium.Marker(
    [closest_coastline["lat"], closest_coastline["lon"]],
    popup="Closest Coastline",
    icon=folium.Icon(color="green", icon="info-sign")
).add_to(site_map)

# Create a distance label marker
distance_marker = folium.Marker(
    [ (launch_site_lat + closest_coastline["lat"]) / 2,  # Midpoint for better display
      (launch_site_lon + closest_coastline["lon"]) / 2 ],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' 
             % "{:10.2f} KM".format(distance_coastline),
    )
)

# Add the distance marker to the map
site_map.add_child(distance_marker)

# Display the map
site_map
```
